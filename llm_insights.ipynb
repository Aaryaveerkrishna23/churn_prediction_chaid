```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data based on the decision tree analysis
data = {
    'Customer Type': ['Month-to-Month', 'One-Year', 'Two-Year'],
    'Total Customers': [1997, 397, 486],
    'Churned Customers': [540, 37, 15],
    'Non-Premium': [1457, 360, 471],
    'Premium': [540, 37, 15]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate churn rates
df['Churn Rate (%)'] = (df['Churned Customers'] / df['Total Customers']) * 100

# Set up the figure
plt.figure(figsize=(15, 12))

# Bar plot for churn rates by customer type
plt.subplot(2, 2, 1)
sns.barplot(x='Customer Type', y='Churn Rate (%)', data=df, palette='Set1')
plt.title('Churn Rate by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Churn Rate (%)')

# Pie chart for payment method distribution (sample data)
payment_data = {
    'Payment Method': ['Bank Transfer (Automatic)', 'Credit Card (Automatic)', 'Electronic Check'],
    'Churn Rate (%)': [15, 45, 40]  # Example churn rates for payment methods
}

payment_df = pd.DataFrame(payment_data)

plt.subplot(2, 2, 2)
plt.pie(payment_df['Churn Rate (%)'], labels=payment_df['Payment Method'], autopct='%1.1f%%', startangle=90)
plt.title('Churn Rate by Payment Method')

# Bar plot for premium vs non-premium churn
plt.subplot(2, 2, 3)
sns.barplot(x=['Non-Premium', 'Premium'], y=[df['Non-Premium'].sum(), df['Premium'].sum()], palette='Set2')
plt.title('Total Customers: Premium vs Non-Premium')
plt.ylabel('Total Customers')

# Churn distribution
plt.subplot(2, 2, 4)
sns.barplot(x='Customer Type', y='Churned Customers', data=df, palette='Set3')
plt.title('Total Churned Customers by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Churned Customers')

# Layout adjustment
plt.tight_layout()
plt.show()
```
